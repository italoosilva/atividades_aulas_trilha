## Questão 6 📝

Crie um programa que recebe o salário de um funcionário e calcula o imposto devido com base nas faixas abaixo:

| Faixa salarial | Alíquota |
|---|---|
| Até R\$ 1.500 | Isento (0%) |
| R\$ 1.500,01 – R\$ 3.000 | 7,5% |
| R\$ 3.000,01 – R\$ 4.500 | 15% |
| Acima de R\$ 4.500 | 22,5% |

Exiba o salário bruto, o valor do imposto e o salário líquido.

````python
salario_bruto = float(input("Digite o valor do salário do funcionário: "))

if salario_bruto <= 1500:
  imposto = "Isento"
  salario_liquido = salario_bruto

elif salario_bruto <= 3000:
  imposto = (salario_bruto - 1500) * 0.075
  imposto = round(imposto, 2)
  salario_liquido = salario_bruto - imposto

elif salario_bruto <= 4500:
  imposto = 1500 * 0.075 + (salario_bruto - 3000) * 0.15
  imposto = round(imposto, 2)
  salario_liquido = salario_bruto - imposto

elif salario_bruto > 4500:
  imposto = 1500 * 0.075 + 1500 * 0.15 + (salario_bruto - 4500) * 0.225
  imposto = round(imposto, 2)
  salario_liquido = salario_bruto - imposto

print(f"\nO salário bruto é: {salario_bruto:.2f}")
print(f"O imposto é: {imposto}")
print(f"O salário líquido é: {salario_liquido:.2f}")
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Use `if`/`elif`/`else` para cada faixa. Calcule `imposto = salario * aliquota`, depois `liquido = salario - imposto`. *

## Questão 7 📝

**Regras:**
- Divisível por 4 → bissexto

Exiba `"Ano bissexto"` ou `"Não é bissexto"`.

## Questão 8 📝

Construa um programa que recebe o peso e a altura do usuário, calcula o IMC e classifica o resultado:

| IMC | Classificação |
|---|---|
| IMC < 18,5 | Abaixo do peso |
| 18,5 <= IMC < 25 | Peso normal |
| 25 <= IMC < 30 | Sobrepeso |
| IMC >= 30 | Obesidade |

Exiba o IMC com 2 casas decimais e a classificação.

````python
peso = float(input("Digite o peso em kg: "))
altura = float(input("Digite a altura em metros: "))

imc = peso / (altura**2)

if imc < 18.5:
  print("\nAbaixo do peso")

elif imc < 25:
  print("\nPeso normal")

elif imc < 30:
  print("\nSobrepeso")

else:
  print("\nObesidade")
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*`IMC = peso / altura**2`. Para formatar com 2 casas decimais, use `f"{imc:.2f}"`.*
